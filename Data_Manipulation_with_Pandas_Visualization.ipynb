```python
#Dataframe manipulation
df=pd.read_csv('ny_weather-1.csv')
df.head()
```

```python
df.tail()
```

```python
df.columns
```

```python
#Dimensions of the dataset
df.ndim
```

```python
#no. of rows and columns
df.shape
```

```python
#Accessing elements
# using loc for getting elements from rows index 6 - 11 and columns Temperature ,Humidity, Event

df.loc[6:11,['Temperature','Humidity','Event']]
```

```python
# using iloc for getting elements from rows index 6 - 11 and columns Temperature ,Humidity, Event

df.iloc[6:12,[1,3,7]]
```

```python
#adding column
values=np.arange(30,44)
values
```

```python
df['new_col']=values
df.head()
```

```python
#removing new column
df.drop('new_col',axis=1,inplace=True)
```

```python
df.head()
```

```python
#adding new row
df.loc[14]=['1/2/2018',39,25,67,43,5,8,'sunny']
df
```

```python
# removing rows index 7 $ 14
df.drop([7,14],inplace=True)
df
```

```python
#setting index as date column
df.set_index('Date',inplace=True)
```

```python
#reset
df.reset_index(inplace=True)
```

```python
#filling na with mean values of each column
df.mean(numeric_only=True)
```

```python
df.fillna({'Temperature':34.9,'DewPoint':17.2,'Humidity':49.63,'Sea Level PressureIn':29.98,'VisibilityMiles':9.72,'WindSpeedMPH':6.30},inplace=True)
```

```python
# dropping remaining null values
df.dropna(inplace=True)
```

```python
df
```

```python
# group by event and calculating avg temp
df.groupby('Event')[['Temperature']].mean()
```

```python
# using apply function to convert visibiltymiles into km
df['VisibilityMiles']=df['VisibilityMiles'].apply(lambda x:x*1.6)
```
